#### clase Aplicación móvil bancaria

```python
class Aplicacion_Movil_Bancaria:
    def __init__(self, numeroCuenta, nombreUsuario, saldo, nip, estadoCuenta):
        self.numeroCuenta = numeroCuenta
        self.nombreUsuario = nombreUsuario
        self.saldo = saldo
        self.nip = nip
        self.estadoCuenta = estadoCuenta

    def consultar_saldo(self):
        print("Tu saldo actual es:", self.saldo)

    def depositar_dinero(self, cantidad):
        if cantidad > 0:
            self.saldo += cantidad
            print("Depósito exitoso.")
        else:
            print("Cantidad inválida.")

    def retirar_dinero(self, cantidad):
        if cantidad <= self.saldo:
            self.saldo -= cantidad
            print("Retiro exitoso.")
        else:
            print("Saldo insuficiente.")

    def cambiar_nip(self, nuevo_nip):
        self.nip = nuevo_nip
        print("NIP actualizado correctamente.")

    def mostrar_datos(self):
        print("Número de cuenta:", self.numeroCuenta)
        print("Usuario:", self.nombreUsuario)
        print("Estado:", self.estadoCuenta)


# Crear objeto
usuario = Aplicacion_Movil_Bancaria(12345, "Hugo Fabian", 1000, 1234, "Activa")

# Menú
while True:
    print("\n--- MENÚ BANCA MÓVIL ---")
    print("1. Consultar saldo")
    print("2. Depositar dinero")
    print("3. Retirar dinero")
    print("4. Cambiar NIP")
    print("5. Mostrar datos")
    print("6. Salir")

    opcion = int(input("Selecciona una opción: "))

    if opcion == 1:
        usuario.consultar_saldo()

    elif opcion == 2:
        cantidad = float(input("Ingresa cantidad a depositar: "))
        usuario.depositar_dinero(cantidad)

    elif opcion == 3:
        cantidad = float(input("Ingresa cantidad a retirar: "))
        usuario.retirar_dinero(cantidad)

    elif opcion == 4:
        nuevo_nip = int(input("Ingresa nuevo NIP: "))
        usuario.cambiar_nip(nuevo_nip)

    elif opcion == 5:
        usuario.mostrar_datos()

    elif opcion == 6:
        print("Gracias por usar la aplicación.")
        break

    else:
        print("Opción no válida.")
```
